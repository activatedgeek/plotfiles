```python
#collapse-hide

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=2., style='whitegrid')
```

## Circle

```python
fig, ax = plt.subplots(figsize=(7,7))

theta = np.linspace(0., 2. * np.pi, 200)
r = np.sqrt(2)
  
x = r * np.cos(theta)
y = r * np.sin(theta)
  
ax.plot(x, y)
ax.set(aspect=1., xlabel='x', ylabel='y')

fig.tight_layout()
```
